plot only, will read a csv file an plot in a pyhton window. 
Dont forget to plot the correct line for the correct genome.

```python
from csv import reader
import matplotlib.pyplot as PL
%matplotlib qt 

#infile_handle = open('/Users/u5264546/Applications/Ricemapping/rice_mapping/data/out/chrom_nippchr07.csv')
infile_handle = open('/Users/u5264546/Applications/Ricemapping/rice_mapping/data/out/chrom_9311chr07.csv')

infile = reader(infile_handle)
P = []
score = []
for line in infile:
    if line[2] == "nan":
        continue
    P.append(int(line[1]))
    score.append(float(line[2]))

PL.figure(1,figsize=[10,5])
PL.clf()
PL.plot(P,score,'k.',alpha=0.5)
#PL.vlines(22779508,0,max(score), colors='red', linestyles='dashed')
PL.vlines(20129204,0,max(score), colors='red', linestyles='dashed')

#print repr(score)

#PL.show()

infile_handle.close()
```
